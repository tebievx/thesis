```python
# plot predictions
plt.figure(figsize=(5, 2.5))
plt.plot(np.arange(trajectory_length), multiple_trajectories.T, color='#ff7f0e', linewidth=0.4, alpha=0.2, zorder=1)
plt.plot(np.arange(trajectory_length), true, color='#1f77b4', linewidth=1.5, zorder=2)
plt.scatter(np.arange(trajectory_length), predicted, color='red', s=25, alpha=0.8, zorder=3)
plt.xlabel('Time step')
plt.ylabel('Time series value')
plt.ylim(top=1)
plt.ylim(bottom=0)
custom_lines = [Line2D([0], [0], color='#1f77b4', lw=1.5),
                Line2D([0], [0], color='#ff7f0e', lw=1),
                Line2D([0], [0], marker='o', color='w', label='Scatter', markerfacecolor='red', markersize=8)]
plt.legend(custom_lines, ['True values', 'Noisy trajectories', 'Final predictions'], bbox_to_anchor=(1.05, 1))
plt.savefig(path + 'base_algorithm.pdf', bbox_inches='tight')
plt.show()
```
